```python
import numpy as np
A = np.array([6,24,60,12,18])
B = np.array([6,4,10,5,4])
print(np.add(A,B)) #adding two arrays
print(np.subtract(A,B))#subtracting two arrays (first -second)
print(np.divide(A,B)) # dividing two arrays
print(np.mod(A,B)) #modulus of two arrays
print(np.divmod(A,B)) # both diving and modulus of two arrays are done first divide array and next modulus array
```
