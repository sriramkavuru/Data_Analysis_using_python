```python
#List Comprehension
#newlist = [output for]
l1=[2,4,6,8,10,12,14,16,18,10]
l2=[i*i for i in l1]
print(l2)
```

```python
#List Comprehension
#newlist = [output for condition]
l3=[23,67,12,34,69]
l4=[i*i for i in l3 if i>50]
print(l4)
```

```python
#newlist = [(st of if)if (cond) else(st of else )for ]
l4=[i*i if i>50 else 'X' for i in l3]
print(l4)
```
